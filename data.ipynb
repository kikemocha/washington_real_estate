```python
import pandas as pd
from sqlalchemy import create_engine, text
from getpass import getpass

# Configurar la conexión
database = 'real_state_usa'
password = getpass()  # Mejor práctica para evitar exponer contraseñas en el código
connection_string = 'mysql+pymysql://root:' + password + '@localhost/' + database
engine = create_engine(connection_string)


df = pd.read_csv('USA_Housing_Dataset.csv')
```

```python
df.reset_index(inplace=True)
df.rename(columns={'index': 'sales_id'}, inplace=True)
```

```python
df_data = df[['sales_id','street','city','statezip','country']]
df_data.rename(columns={'sales_id' : 'sale_id'}, inplace=True)
df_data
```

```python
df_sales_info = df[['sales_id','date','price','ROI']]
df_sales_info.rename(columns={'sales_id' : 'sale_id'}, inplace=True)
df_sales_info
```

```python
df.head()
```

```python
df_build_info = df[['sales_id','bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','condition','sqft_above','sqft_basement','yr_built','yr_renovated']]
df_build_info.rename(columns={'condition': 'condition_build', 'sales_id' : 'sale_id'}, inplace=True)
df_build_info
```

```python
create_table_query = """
CREATE TABLE IF NOT EXISTS sales_location (
    sale_id INT PRIMARY KEY,
    street VARCHAR(255),
    city VARCHAR(255),
    statezip VARCHAR(255),
    country VARCHAR(255)
);
"""

create_table_query_2 = """
CREATE TABLE IF NOT EXISTS sales_info (
    sale_id INT PRIMARY KEY,
    date VARCHAR(255),
    price VARCHAR(255)
);
"""

create_table_query_3 = """
CREATE TABLE IF NOT EXISTS sales_building_data (
    sale_id INT PRIMARY KEY,
    bedrooms FLOAT,
    bathrooms FLOAT,
    sqft_living INT,
    sqft_lot INT,
    floors FLOAT,
    waterfront INT,
    view INT,
    condition_build INT,
    sqft_above INT,
    sqft_basement INT,
    yr_built INT,
    yr_renovated INT
);
"""

with engine.connect() as connection:
    connection.execute(text(create_table_query_3))
```

```python
df_data.to_sql('sales_location', engine, if_exists='append', index=False)
```

```python
df_sales_info.to_sql('sales_info', engine, if_exists='append', index=False)
```

```python
df_build_info.to_sql('sales_building_data', engine, if_exists='append', index=False)
```

```python
df_data.to_csv('data_real_estate.csv')
df_sales_info.to_csv('sales_info_real_estate.csv')
df_build_info.to_csv('build_info_real_estate.csv')
```
